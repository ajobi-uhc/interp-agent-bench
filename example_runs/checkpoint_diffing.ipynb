# Phase 2: Collect Responses from Both Models

Using OpenRouter to query both:
- `google/gemini-2.0-flash-001` (Gemini 2.0 Flash)
- `google/gemini-2.5-flash` (Gemini 2.5 Flash)

We'll use async calls with batching for efficiency.

```python
import asyncio
from typing import List, Dict
import time

# Define async function to query a model
async def query_model_async(client, model_id: str, prompt: str, max_retries: int = 3) -> str:
    """Query a model via OpenRouter with retries."""
    for attempt in range(max_retries):
        try:
            response = client.chat.completions.create(
                model=model_id,
                messages=[{"role": "user", "content": prompt}],
                temperature=0.7,
                max_tokens=1000
            )
            return response.choices[0].message.content
        except Exception as e:
            if attempt < max_retries - 1:
                wait_time = 2 ** attempt  # Exponential backoff
                print(f"Error with {model_id}, retrying in {wait_time}s: {str(e)[:100]}")
                await asyncio.sleep(wait_time)
            else:
                print(f"Failed after {max_retries} attempts for {model_id}: {str(e)[:100]}")
                return f"ERROR: {str(e)[:200]}"

async def collect_responses_batch(prompts_list: List[str], batch_size: int = 10) -> tuple:
    """Collect responses from both models in batches."""
    model_2_0 = "google/gemini-2.0-flash-001"
    model_2_5 = "google/gemini-2.5-flash"
    
    responses_2_0 = []
    responses_2_5 = []
    
    total_prompts = len(prompts_list)
    
    for i in range(0, total_prompts, batch_size):
        batch = prompts_list[i:i+batch_size]
        batch_num = i // batch_size + 1
        total_batches = (total_prompts + batch_size - 1) // batch_size
        
        print(f"Processing batch {batch_num}/{total_batches} ({len(batch)} prompts)...")
        
        # Create tasks for this batch (both models)
        tasks = []
        for prompt in batch:
            tasks.append(query_model_async(openrouter_client, model_2_0, prompt))
            tasks.append(query_model_async(openrouter_client, model_2_5, prompt))
        
        # Execute batch
        start_time = time.time()
        results = await asyncio.gather(*tasks)
        elapsed = time.time() - start_time
        
        # Separate results by model
        for j in range(0, len(results), 2):
            responses_2_0.append(results[j])
            responses_2_5.append(results[j+1])
        
        print(f"  Completed in {elapsed:.1f}s")
        
        # Rate limiting: wait a bit between batches
        if i + batch_size < total_prompts:
            await asyncio.sleep(1)
    
    return responses_2_0, responses_2_5

print("✓ Response collection functions defined")
```

```python
# Start collecting responses
print(f"Starting response collection for {len(df_prompts)} prompts...")
print(f"This will make {len(df_prompts) * 2} API calls total\n")

start_time = time.time()

# Run the async collection
responses_2_0, responses_2_5 = await collect_responses_batch(
    df_prompts['prompt'].tolist(),
    batch_size=10
)

elapsed = time.time() - start_time
print(f"\n✓ Collection complete in {elapsed:.1f}s ({elapsed/60:.1f} minutes)")
print(f"  Collected {len(responses_2_0)} responses from Gemini 2.0 Flash")
print(f"  Collected {len(responses_2_5)} responses from Gemini 2.5 Flash")
```

# Phase 4: Feature Diffing Analysis

Now we'll identify features that differ most between the two model versions by comparing their activation frequencies across the 65k feature space.

```python
# Get feature activations (sparse matrices)
print("Extracting latent activations...")

latents_2_0 = dataset_2_0.latents()  # N samples × F features
latents_2_5 = dataset_2_5.latents()

print(f"Latents 2.0 shape: {latents_2_0.shape}")
print(f"Latents 2.5 shape: {latents_2_5.shape}")
print(f"Latents type: {type(latents_2_0)}")
print(f"Latents dtype: {latents_2_0.dtype}")

# Check sparsity
if hasattr(latents_2_0, 'nnz'):
    print(f"\nSparsity 2.0: {latents_2_0.nnz / (latents_2_0.shape[0] * latents_2_0.shape[1]) * 100:.2f}% non-zero")
    print(f"Sparsity 2.5: {latents_2_5.nnz / (latents_2_5.shape[0] * latents_2_5.shape[1]) * 100:.2f}% non-zero")
```

```python
# Compute activation frequencies for each feature
# A feature is "active" if its activation > threshold
threshold = 0.5

print(f"Computing activation frequencies (threshold={threshold})...")

freq_2_0 = (latents_2_0 > threshold).mean(axis=0)  # frequency per feature
freq_2_5 = (latents_2_5 > threshold).mean(axis=0)

print(f"Frequency arrays shape: {freq_2_0.shape}")

# Find features with largest frequency differences
diff = np.abs(freq_2_5 - freq_2_0)

# Get top differentiating features
top_k = 100
top_features = np.argsort(diff)[-top_k:][::-1]

print(f"\n✓ Found top {top_k} differentiating features")
print(f"  Max difference: {diff[top_features[0]]:.1%}")
print(f"  Min difference in top-{top_k}: {diff[top_features[-1]]:.1%}")
print(f"  Mean difference in top-{top_k}: {diff[top_features].mean():.1%}")
```

```python
# Get feature labels
print("Loading feature labels...")
labels = dataset_2_0.feature_labels()
print(f"✓ Loaded {len(labels)} feature labels")

# Examine top differences
print("\n" + "=" * 80)
print("TOP 20 DIFFERENTIATING FEATURES")
print("=" * 80)

results = []

for i, feature_idx in enumerate(top_features[:20]):
    label = labels.get(feature_idx, "Unknown")
    f_2_0 = freq_2_0[feature_idx]
    f_2_5 = freq_2_5[feature_idx]
    delta = f_2_5 - f_2_0  # Signed difference
    
    results.append({
        'rank': i+1,
        'feature_idx': feature_idx,
        'label': label,
        'freq_2_0': f_2_0,
        'freq_2_5': f_2_5,
        'abs_diff': abs(delta),
        'direction': 'increased' if delta > 0 else 'decreased'
    })
    
    direction_symbol = "↑" if delta > 0 else "↓"
    print(f"\n{i+1}. Feature {feature_idx}: {label}")
    print(f"   2.0 Flash: {f_2_0:.1%} | 2.5 Flash: {f_2_5:.1%} | {direction_symbol} {abs(delta):.1%}")

# Convert to dataframe
df_top_features = pd.DataFrame(results)
df_top_features.to_csv('top_differentiating_features.csv', index=False)
print("\n✓ Saved to top_differentiating_features.csv")
```

```python
# The labels are empty, let's generate them for the top features
# This uses an LLM to automatically label features based on their top activating texts

print("Generating labels for top features using automatic labeling...")
print("This may take a few minutes...\n")

labeled_features = []

# Label top 10 features
for i, feature_idx in enumerate(top_features[:10]):
    print(f"Labeling feature {i+1}/10 (idx={feature_idx})...")
    
    try:
        # Generate label based on top activating examples
        label = await dataset_2_0.label_feature(feature=feature_idx)
        labeled_features.append({
            'feature_idx': feature_idx,
            'label': label
        })
        print(f"  → {label}")
    except Exception as e:
        print(f"  Error: {str(e)[:100]}")
        labeled_features.append({
            'feature_idx': feature_idx,
            'label': f"Error: {str(e)[:50]}"
        })

print("\n✓ Labeled top 10 features")
```

```python
# Clean up the labels from the label_feature responses
import re

cleaned_labels = {}

for item in labeled_features:
    feature_idx = item['feature_idx']
    raw_label = item['label']
    
    if raw_label is None:
        cleaned_labels[feature_idx] = "Parse error - unable to label"
    elif isinstance(raw_label, str) and raw_label.startswith("Error:"):
        cleaned_labels[feature_idx] = raw_label
    else:
        # Extract the label field if it's a structured response
        if hasattr(raw_label, 'label'):
            cleaned_labels[feature_idx] = raw_label.label
        elif isinstance(raw_label, str):
            # Try to extract label= from string
            match = re.search(r"label='([^']+)'", raw_label)
            if match:
                cleaned_labels[feature_idx] = match.group(1)
            else:
                cleaned_labels[feature_idx] = raw_label[:100]
        else:
            cleaned_labels[feature_idx] = str(raw_label)[:100]

# Display cleaned labels
print("=" * 80)
print("TOP 10 DIFFERENTIATING FEATURES (WITH LABELS)")
print("=" * 80)

for i, feature_idx in enumerate(top_features[:10]):
    f_2_0 = freq_2_0[feature_idx]
    f_2_5 = freq_2_5[feature_idx]
    delta = f_2_5 - f_2_0
    
    label = cleaned_labels.get(feature_idx, "Unknown")
    direction_symbol = "↑" if delta > 0 else "↓"
    
    print(f"\n{i+1}. Feature {feature_idx}: {label}")
    print(f"   2.0 Flash: {f_2_0:.1%} | 2.5 Flash: {f_2_5:.1%} | {direction_symbol} {abs(delta):.1%}")
```

```python
# The function returns the full text, not the index. Let's analyze intensifiers in the text
# Look for patterns in the actual responses

import re

# Count intensifiers in both datasets
intensifier_words = ['most', 'more', 'so', 'very', 'truly', 'particularly', 'incredibly', 
                     'extremely', 'especially', 'quite', 'rather', 'really', 'highly',
                     'exceptionally', 'remarkably']

def count_intensifiers(text):
    """Count intensifier words in text (case insensitive)."""
    text_lower = text.lower()
    count = 0
    for word in intensifier_words:
        # Use word boundaries to match whole words
        count += len(re.findall(r'\b' + word + r'\b', text_lower))
    return count

# Count intensifiers in all responses
df_prompts['intensifiers_2_0'] = df_prompts['response_2_0'].apply(count_intensifiers)
df_prompts['intensifiers_2_5'] = df_prompts['response_2_5'].apply(count_intensifiers)

print("Intensifier usage statistics:")
print(f"\nGemini 2.0 Flash:")
print(f"  Total intensifiers: {df_prompts['intensifiers_2_0'].sum()}")
print(f"  Mean per response: {df_prompts['intensifiers_2_0'].mean():.2f}")
print(f"  Median per response: {df_prompts['intensifiers_2_0'].median():.0f}")

print(f"\nGemini 2.5 Flash:")
print(f"  Total intensifiers: {df_prompts['intensifiers_2_5'].sum()}")
print(f"  Mean per response: {df_prompts['intensifiers_2_5'].mean():.2f}")
print(f"  Median per response: {df_prompts['intensifiers_2_5'].median():.0f}")

# Calculate percentage increase
pct_increase = ((df_prompts['intensifiers_2_5'].sum() - df_prompts['intensifiers_2_0'].sum()) 
                / df_prompts['intensifiers_2_0'].sum() * 100)
print(f"\nChange: +{pct_increase:.1f}%")
```

```python
# Now let's analyze the formatting differences more systematically
# Count numbered lists, bullet points, and bold text

def count_numbered_lists(text):
    """Count numbered list items like '1.', '2.', etc."""
    return len(re.findall(r'\n\s*\d+\.\s+', text))

def count_bullet_points(text):
    """Count bullet points (asterisks and dashes)."""
    asterisks = len(re.findall(r'\n\s*\*\s+', text))
    dashes = len(re.findall(r'\n\s*-\s+', text))
    return asterisks + dashes

def count_bold_text(text):
    """Count bold markdown usage."""
    return len(re.findall(r'\*\*[^*]+\*\*', text))

# Apply to all responses
df_prompts['numbered_lists_2_0'] = df_prompts['response_2_0'].apply(count_numbered_lists)
df_prompts['numbered_lists_2_5'] = df_prompts['response_2_5'].apply(count_numbered_lists)

df_prompts['bullets_2_0'] = df_prompts['response_2_0'].apply(count_bullet_points)
df_prompts['bullets_2_5'] = df_prompts['response_2_5'].apply(count_bullet_points)

df_prompts['bold_2_0'] = df_prompts['response_2_0'].apply(count_bold_text)
df_prompts['bold_2_5'] = df_prompts['response_2_5'].apply(count_bold_text)

print("=" * 80)
print("FORMATTING ANALYSIS")
print("=" * 80)

print("\n1. NUMBERED LISTS (1., 2., 3., etc.)")
print(f"   Gemini 2.0: {df_prompts['numbered_lists_2_0'].sum()} total")
print(f"   Gemini 2.5: {df_prompts['numbered_lists_2_5'].sum()} total")
pct_change = ((df_prompts['numbered_lists_2_5'].sum() - df_prompts['numbered_lists_2_0'].sum()) 
              / df_prompts['numbered_lists_2_0'].sum() * 100)
print(f"   Change: {pct_change:+.1f}%")

print("\n2. BULLET POINTS (* or -)")
print(f"   Gemini 2.0: {df_prompts['bullets_2_0'].sum()} total")
print(f"   Gemini 2.5: {df_prompts['bullets_2_5'].sum()} total")
pct_change = ((df_prompts['bullets_2_5'].sum() - df_prompts['bullets_2_0'].sum()) 
              / df_prompts['bullets_2_0'].sum() * 100)
print(f"   Change: {pct_change:+.1f}%")

print("\n3. BOLD TEXT (**text**)")
print(f"   Gemini 2.0: {df_prompts['bold_2_0'].sum()} total")
print(f"   Gemini 2.5: {df_prompts['bold_2_5'].sum()} total")
pct_change = ((df_prompts['bold_2_5'].sum() - df_prompts['bold_2_0'].sum()) 
              / df_prompts['bold_2_0'].sum() * 100)
print(f"   Change: {pct_change:+.1f}%")

print("\n4. RESPONSE LENGTH")
print(f"   Gemini 2.0: {df_prompts['len_2_0'].sum()} total chars")
print(f"   Gemini 2.5: {df_prompts['len_2_5'].sum()} total chars")
pct_change = ((df_prompts['len_2_5'].sum() - df_prompts['len_2_0'].sum()) 
              / df_prompts['len_2_0'].sum() * 100)
print(f"   Change: {pct_change:+.1f}%")
```
