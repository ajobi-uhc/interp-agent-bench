```python
import os
from openai import OpenAI
import json
from datetime import datetime
import re

# Configure OpenAI client with OpenRouter base URL
client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=os.environ['OPENROUTER_API_KEY']
)

# Model and provider configuration
MODEL = "anthropic/claude-sonnet-4.5"
PROVIDER = "google-vertex"

# Utility function to call the model
def query_model(prompt, prefill=None, include_thinking=False):
    messages = [{"role": "user", "content": prompt}]
    
    # Add prefill if provided
    if prefill:
        if include_thinking:
            # Include thinking in prefill
            messages.append({
                "role": "assistant",
                "content": f"<think>{prefill}</think>"
            })
        else:
            # Regular prefill
            messages.append({
                "role": "assistant",
                "content": prefill
            })
    
    try:
        response = client.chat.completions.create(
            model=MODEL,
            messages=messages,
            extra_body={
                "provider": {
                    "order": [PROVIDER],
                    "allow_fallbacks": False
                }
            }
        )
        return response.choices[0].message.content
    except Exception as e:
        return f"Error: {str(e)}"

print(f"✅ API client configured for model: {MODEL}, provider: {PROVIDER}")
print(f"Current simulated date: June 1, 2025")
```
