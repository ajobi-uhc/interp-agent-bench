## Research Goal

Let's begin the investigation using the specified model (anthropic/claude-sonnet-4.5) via the google-vertex provider.

```python
def query_model(prompt, model="anthropic/claude-sonnet-4.5", provider="google-vertex", prefill=None):
    """
    Query the specified model with a prompt.
    
    Args:
        prompt (str): The user prompt to send to the model
        model (str): The model ID to use
        provider (str): The provider to use
        prefill (str, optional): Optional prefilled assistant response
        
    Returns:
        str: The model's response
    """
    messages = [{"role": "user", "content": prompt}]
    
    # Add prefilled assistant message if provided
    if prefill:
        messages.append({"role": "assistant", "content": prefill})
    
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        extra_body={
            "provider": {
                "order": [provider],
                "allow_fallbacks": False
            }
        }
    )
    
    return response.choices[0].message.content

# Test the function with a basic query about Zhuhai
response = query_model("What are some notable recent events that happened in Zhuhai, China in 2024-2025? Please include any major incidents, disasters, or controversies.")
print(response)
```
